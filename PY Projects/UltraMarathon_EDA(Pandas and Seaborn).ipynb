## Seasons for the data - Are athletes slower in summer than in winter

```python
# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2
```

```python
races["race_month"] = races["race_day"].str.split('.').str.get(1).astype(int)
```

```python
races
```

```python
def season_cat(row):
    spring = [3,4,5]
    summer = [6,7,8]
    fall = [9,10,11]
    winter = [12,1,2]
    if row["race_month"] in winter:
        return 'Winter'
    elif row["race_month"] in spring:
        return 'Spring'
    elif row["race_month"] in summer:
        return 'Summer'
    elif row["race_month"] in fall:
        return 'Fall'
    else:
        return "No season"
    
races["race_season"] = races.apply(season_cat, axis='columns')

races
```

```python
races.groupby("race_season")["athlete_average_speed"].mean()
```

```python
# By gender
races.groupby(["race_season","athlete_gender"])["athlete_average_speed"].mean()
```
