# Setting Up

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
reviews = pd.read_csv("Downloads\winemag-data-130k-v2.csv\winemag-data-130k-v2.csv",index_col=0)
# pd.set_option('display.max_colwidth', None)
```

```python
reviews.head()
```

```python
reviews.shape
```

# Data Cleaning and Value Checks

```python
reviews.describe()
```

```python
reviews.info()
```

```python
reviews["country"].fillna("Unknown", inplace=True)
reviews
```

```python
reviews["designation"].fillna("Unknown", inplace=True)
reviews
```

```python
price_mean = reviews["price"].mean()
reviews["price"].fillna(price_mean, inplace=True)
reviews
```

```python
reviews["province"].fillna("Unknown", inplace=True)
reviews
```

```python
reviews["region_1"].fillna("Unknown", inplace=True)
reviews
```

```python
reviews.drop(columns="taster_twitter_handle", inplace=True)
```

```python
reviews.drop_duplicates(inplace=True)
```

```python
reviews.head()
```

# Data Visualisations

## What is the distribution of wine prices in the dataset?

```python
price_dist = reviews.groupby("price").size().reset_index(name="count")
price_dist
```

```python
plt.figure(figsize=(16,8))
sns.histplot(x="price", bins=30, data=price_dist)
```

## Distribution of wine ratings (points).

```python
plt.figure(figsize=(10,6))
sns.histplot(reviews['points'], binwidth=1, kde=True, color="skyblue")
plt.title('Distribution of Wine Points')
plt.xlabel('Points')
plt.ylabel('Frequency')
```

## Generate a box plot to compare the prices of wines produced in different countries.

```python
plt.figure(figsize=(20,10))
sns.boxplot(x="country", y="price", data=reviews)
plt.title("Prices of wines produced in different countries")
plt.xlabel("Country")
plt.ylabel("Price")
plt.xticks(rotation=45, ha='right');
```

## Top 10 countries with the highest average wine prices.

```python
top10 = reviews.groupby("country")["price"].mean().sort_values(ascending=False).reset_index().head(10)
top10
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="country", y="price", data=top10, palette="pastel")
plt.title("Top 10 Countries By Highest Average Price")
plt.xlabel("Country")
plt.ylabel("Average Price")
plt.xticks(rotation=45, ha='right');
```

## Is there a relationship between wine ratings (points) and prices

```python
sns.regplot(x="points", y="price", data=reviews)
```

## Distribution of wine varieties.

```python
top_variety = reviews.groupby("variety").size().sort_values(ascending=False).reset_index(name="count").head(50)
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(x="variety", y="count", data=top_variety)
plt.title('Distribution of Wine Varieties')
plt.xlabel('Wine Variety')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right');
```

## Wine prices by taster

```python
plt.figure(figsize=(16,8))
sns.boxplot(x="taster_name", y="price", data=reviews)
plt.title("Prices of Wines By Tasters")
plt.xlabel("Taster")
plt.ylabel("Price")
plt.xticks(rotation=45, ha='right');
```

## Heatmap to explore the correlation between wine ratings and prices.

```python
correlations = reviews.corr()
correlations
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data=correlations, annot=True)
plt.title('Correlation between Wine Ratings and Prices.')
```

## Countries with the best rated wines

```python
best_rated = reviews.groupby("country")["points"].mean().sort_values(ascending=False).reset_index()
best_rated
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x="country",y="points", data=best_rated, palette="pastel")
plt.title("Country Rankings Based On Scores")
plt.xlabel("Country")
plt.ylabel("Points")
plt.xticks(rotation=45, ha='right');
```
